```python
import pandas as pd

data = pd.read_csv('train.csv')
data.head()
```

```python
data.Name.tolist()
```

```python
# Clean / transform data
# First, which of the columns are either all unique / all the same

cols = data.columns.tolist()
print(f"num rows: {len(data)}")
for col in cols:

    print(f'Column: {col}, {len(data[col].unique())}')
```

```python
# So passengerId and Name have no bearing on survivability
data = data.drop(columns=['PassengerId', 'Name'])
```

```python
# Check out NaN's
for col in data.columns:
    print(f'Column: {col}, {data[col].isna().sum()}')
```

```python
# So Age is continuous; Cabin and Embarked are categorical. 
# Let's just replace age with average age, and Cabin/Embarked nan's with Unk's. 

avg_age = data.Age.mean()
data.Age = data.Age.fillna(avg_age)

data.Cabin = data.Cabin.fillna("Unknown")
data.Embarked = data.Embarked.fillna("Unknown")
```

```python
data = pd.get_dummies(data, columns=['Pclass', 'Cabin', 'Embarked'])
data
```

```python
# Replace sex
sex = {'male': 1,'female': 0}

data.Sex = [sex[item] for item in data.Sex]
```

```python
#drop Ticket
data = data.drop(['Ticket'], axis=1)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

X = data.drop(columns=['Survived']).to_numpy()
y = data.Survived.to_numpy()


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, 
                    hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(X_train, y_train)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
preds = clf.predict(X_test)

accuracy_score(y_test, preds)
```
